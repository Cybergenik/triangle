# Constitution & Resolutions Guide

This should serve as a general guideline for how to write in the same style as a Constitutional clause or Resolution.

## Constitutional Writing

The Constitution for our Chapter is meant to define a skeleton structure for how we operate. There are, generally, three Articles to the Constitution:

* **Cause**, which defines the core purpose and pursuable future vision for the Chapter. A good analog would be the Preamble to the United States Declaration of Independence. In the same way the Declaration declare the vision of a society that works to defend inaliable rights, our Cause should declare the vision of a chapter that works to enforce and build upon both the values of Triangle Fraternity and the values of our Chapter.
* **Membership** defines the commitments that we, as Brothers, commit ourselves to by virtue of our Brotherhood. It defines the same commitments for New Members and Alumni Brothers.
* **Organization** defines the horizontal stratification of our Chapter into a set of legsilative, executive and judicial systems. It defines the powers, authorities, and commitments of each body and the Brothers who operate that system.

The Constitution is meant to be a more concrete, core set of definitions of our Chapter's system. It should only define things that are unlikely to be subject to wide variation between generations. Because of this, some of the language can sound vague. The intent is that the skeleton of our Chapter is built by the Constitution, while the primary movements of the Chapter are established by Resolutions.

An example of this would be Chapter Meeting attendence requirements. Because this has been subject to variable change, the Constitution itself only defines that:

> Each Brother of our Chapter's Legislature shall commit himself to regularly attend and be a constructive participant at Chapter Meetings.

The commitment is defined here - the exact details by which that commitment is fulfilled is defined by a Resolution:

> The members of the Legislature of our Chapter shall resolve to define fulfillment of their commitment to regularly attend Chapter Meetings as attendance at **50%** of standard Chapter Meetings.

### Clause Structure

In general, clauses within the Constitution act as a definition on how some Chapter matter should operate. should follow the format of:

> [PARTY] shall [OPERATION] [SUBJECT]

where:

* [PARTY] is the person or group of people tasked with executing the [OPERATION];
* [OPERATION] is whatever the [PARTY] is supposed to do; and
* [SUBJECT] is the matter that the [OPERATION] applies to and the [PARTY] is tasked with carrying out.

For example, here is the clause defining how outstanding academic achievement should be handled:

> Our Chapter shall recognize, each term, each Brother with a cumulative or term GPA at **4.0** for his outstanding academic achievement, and shall record his distinctive and honorable performance in our Chapter's history.

This clause can be broken into the following components:

* [PARTY] Our Chapter
* [OPERATION] shall recognize, each term,
* [SUBJECT] each Brother with a cumulative or term GPA at **4.0** for his outstanding academic achievement, and
* [OPERATION] shall record
* [SUBJECT] his (the Brother's) distinctive and honorable performance in our Chapter's history.

This clearly defines who needs to act, what they need to do, and how they should go about doing it. Good Constitutional definitions should leave some of the matter to interpretation. In this case, how exactly the Chapter should go about recognizing and recording the academic achievement - this can be left to the Chapter to decide.

Clauses may consist of multiple sentences - each Clause, though, should tend to cover one topic or matter. For example, here is the entire clause for the establishment of Meeting minutes.

> The Brother Secretary of our Chapter shall lead the minuting of the Overview and discussions of New Business of each meeting. Chapter Meeting minutes shall reflect, in good faith, the information of the Overview and discussions of New Business. The Brother Secretary shall ensure Chapter Meeting minutes are available following the minuted Chapter Meeting.

This clause can be broken into the following components: 

* [PARTY] The Brother Secretary of our Chapter
* [OPERATION] shall lead
* [SUBJECT] the minuting of the Overview and discussions of New Business in each meeting.

<hr>

* [PARTY] Chapter Meeting minutes
* [OPERATION] shall reflect, in good faith,
* [SUBJECT] the information of the Overview and discussions of New Business.

<hr>

* [PARTY] The Brother Secretary of our Chapter
* [OPERATION] shall ensure
* [SUBJECT] Chapter Meeting minutes are available following the minuted Chapter Meeting.

In this case, we have three distinct sub-clauses within this clause. Three sub-clauses should generally be considered a sfot maximum for a composite clause - even in this instance, the clause could stand to be split.

Additionally, certan aspects of this topic are left vague - how soon after the meeting should minutes be expected to be available, for example? Exact details should be left to a Resolution - for example, a possible sister Resolution could be:

> Our Chapter resolves to make Chapter meeting minutes available no later than a week following the minuted Meeting.

## Resolution Writing

Resolutions are meant to provide a more detailed interpretation of the underlying structure put forth in the Constitution.

### Resolution Structure

Resolutions follow a similar structure to Constitutional definitions, where each resolution is structured as:

* [PARTY] shall resolve to [TASK]

where:

* [PARTY] is the party that makes the resolution to do something (or where a resolution is made for them by the Chapter; and
* [TASK] is exactly what the [PARTY] resolves to do; and, optionally
* [EXECUTOR] is who is supposed to see out the specific execution of the [TASK].

Resolutions should, in the vast majority of cases, act as a sister explanation to a Constitutional definition.

An example of this would be a Resolution pulled from the original JAMM amendment. JAMM was, in the course of re-writing the Constitution, split into a core mechanism for meeting structures that remained in the Constitution and a set of Resolutions that define exactly how the structure should be maneuvered.

The core Constitutional definitions: 

> Each Chapter Meeting shall have an expected duration. Brothers and New Members shall commit to attending for the full expected duration of a Chapter Meeting, but shall not be expected to remain past the expected duration of a Chapter Meeting.
>
> [...]
>
> Each Chapter Meeting shall have time allocated for an Overview of any New Business, led by our President and conducted by the Executive Board. Our President and Executive Board shall commit to keeping this Overview to a reasonable time.

This definition is vague - from this definition, we know that a Chapter Meeting should have an Overview and New Business somewhere in there, but not much definition beyond that.

This is given more specific criteria by two sister Resolutions:

> Our Chapter shall resolve to keep the duration of the Overview of a Chapter Meeting to less than **ten** minutes. The President of our Chapter shall execute this Resolution.
>
> Our Chapter shall resolve to keep the duration of standard Chapter Meetings to **one hour,** and expect prior notification if a Chapter Meeting is anticipated to require more than the standard alloted time. The President of our Chapter shall execute this Resolution.
